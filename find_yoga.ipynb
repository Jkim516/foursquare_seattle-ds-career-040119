```python
import json
import requests

# Load secrets from credentials.json
url = 'https://api.foursquare.com/v2/venues/explore'
with open('.secrets/credentials.json') as f:
    params = json.load(f)
```

```python
params['v'] = '20190717'
params['ll'] = '47.608,-122.336',
params['query'] = 'yoga',
params['intent'] = 'browse',
params['radius'] = 100000,
params['limit'] = 100
```

```python
response = requests.get(url=url, params=params)
data = json.loads(response.text)
```

```python
groups = data['response']['groups']
```

```python
items = groups[0]['items']
```

```python
# or make function
def get_name(item):
    return item['venue']['name']

def get_address(item):
    return item['venue']['location'].get('address')
```
